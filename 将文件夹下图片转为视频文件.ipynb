```python
import os
import cv2
import numpy as np

path = './mnist/'
filelist = os.listdir(path)

fps = 24 #24
size = (28, 28) #
#cv2.resize()

video = cv2.VideoWriter("VideoTest1.avi", cv2.VideoWriter_fourcc('I', '4', '2', '0'), fps, size)
#

for item in filelist:
    if item.endswith('.jpg'): 
    #.png.jpg
        item = path + item
        print(item)
        img = cv2.imread(item)
        video.write(img)

video.release()
cv2.destroyAllWindows()
```
